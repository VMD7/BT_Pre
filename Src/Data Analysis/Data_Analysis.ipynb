```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
sns.set_style("whitegrid")
import altair as alt
alt.renderers.enable("notebook")

# Code for hiding seaborn warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Dataset/BT1 Dataset.csv",encoding='latin1')
```

```python
df.head()
```

```python
df['Questions_length'] = df['Questions'].str.len()
```

```python
plt.figure(figsize=(12.8,6))
sns.distplot(df['Questions_length']).set_title('Questions length distribution');
```

```python

df['Questions_length'].describe()
```

```python
quantile_95 = df['Questions_length'].quantile(0.95)
df_95 = df[df['Questions_length'] < quantile_95]
```

```python
plt.figure(figsize=(12.8,6))
sns.distplot(df_95['Questions_length']).set_title('Questions length distribution');
```

```python
df_more10k = df[df['Questions_length'] > 100]
len(df_more10k)
```
